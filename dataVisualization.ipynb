### Visualising the distribution of the training data set

```python
import os
from glob import glob
import numpy as np
import matplotlib.pyplot as plt

data_path = "face_age"
x = np.arange(1, 101)
y = []
z = []

for i in x:
    filenames = glob(os.path.join(data_path, f"{i:03}", '*.[PJpj][NPnp][Gg]'))
    y.append(len(filenames))
    filenames = glob(os.path.join(data_path, f"t{i:03}", '*.[PJpj][NPnp][Gg]'))
    z.append(len(filenames))


fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(11,4))

    
ax1.bar(x, y)
ax1.set_xlabel("Age")
ax1.set_ylabel("Occurences")
ax1.set_title("Distribution of ages in training dataset")




ax2.bar(x, z)
ax2.set_xlabel("Age")

ax2.set_title("Distribution of ages in test dataset")
fig.tight_layout()
```
