### Visualising the distribution of the training data set

```python
import os
from glob import glob
import numpy as np
import matplotlib.pyplot as plt

data_path = "face_age"
x = np.arange(1, 101)
y = []

for i in x:
    filenames = glob(os.path.join(data_path, f"{i:03}", '*.[PJpj][NPnp][Gg]'))
    y.append(len(filenames))

plt.bar(x, y)
plt.xlabel("Age")
plt.ylabel("Occurence")
plt.title("Distribution of ages in training dataset")
plt.show()
```

### Visualising the testing dataset
